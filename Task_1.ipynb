```python
import pandas as pd
import numpy as np

# 1) Load CSV
data = pd.read_csv("loan_data.csv")
print(data)

# 2) Split into attributes and target
d = np.array(data)[:, :-1]
target = np.array(data)[:, -1]
print("\nThe attributes are:\n", d)
print("\nThe target is:\n", target)

# 3) Find-S training
def train(c, t):
    # take first positive example
    for i, val in enumerate(t):
        if val == "Yes":
            specific_hypothesis = c[i].copy()
            break

    # generalize with remaining positive examples
    for i, val in enumerate(c):
        if t[i] == "Yes":
            for x in range(len(specific_hypothesis)):
                if val[x] != specific_hypothesis[x]:
                    specific_hypothesis[x] = "?"
    return specific_hypothesis

# 4) Final output
print("\nThe final hypothesis is:", train(d, target))
```

```python
import pandas as pd
import numpy as np

# Load CSV
data = pd.read_csv("loan_data.csv")
print(data)

# Split into attributes and target
X = np.array(data.iloc[:, 0:-1])
y = np.array(data.iloc[:, -1])

print("\nThe attributes are:\n", X)
print("\nThe target is:\n", y)

def find_s(X, y):
    # Initialize with first positive example
    h = None
    for i in range(len(y)):
        if y[i].strip().lower() == "yes":
            h = X[i].copy()
            break

    if h is None:
        return None  # no positive examples

    # Generalize using only positive examples
    for i in range(len(y)):
        if y[i].strip().lower() == "yes":
            for j in range(len(h)):
                if X[i][j] != h[j]:
                    h[j] = "?"
    return h

final_h = find_s(X, y)
print("\nThe final hypothesis is:", final_h)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("loan_data.csv")
X = np.array(data.iloc[:, 0:-1])
y = np.array(data.iloc[:, -1])

def covers(h, x):
    # h covers x if each attribute matches or is '?'
    return all(h[i] == "?" or h[i] == x[i] for i in range(len(h)))

def more_general_or_equal(g, s):
    # g is more general than or equal to s
    return all(g[i] == "?" or g[i] == s[i] for i in range(len(g)))

def candidate_elimination(X, y):
    n = X.shape[1]

    # S: most specific hypothesis (use None markers initially)
    S = [None] * n
    # G: set with the most general hypothesis
    G = [["?"] * n]

    for i in range(len(X)):
        x = X[i]
        label = y[i].strip().lower()

        if label == "yes":
            # 1) remove from G any hypothesis that does NOT cover x
            G = [g for g in G if covers(g, x)]

            # 2) generalize S minimally to cover x
            if S == [None] * n:
                S = list(x)
            else:
                for j in range(n):
                    if S[j] is None:
                        S[j] = x[j]
                    elif S[j] != x[j]:
                        S[j] = "?"

        elif label == "no":
            # 1) if S not initialized, skip (or keep as is)
            if S == [None] * n:
                continue

            # 2) specialize G minimally so it does not cover x
            new_G = []
            for g in G:
                if covers(g, x):
                    for j in range(n):
                        # specialize only where S has a specific value different from x
                        if S[j] != "?" and S[j] != x[j]:
                            g_new = g.copy()
                            g_new[j] = S[j]
                            new_G.append(g_new)
                else:
                    new_G.append(g)

            # remove duplicates
            G = []
            for g in new_G:
                if g not in G:
                    G.append(g)

            # keep only hypotheses that are >= S (consistent boundary)
            G = [g for g in G if more_general_or_equal(g, S)]

        print(f"After instance {i+1} ({y[i]}):")
        print("S =", S)
        print("G =", G)
        print("-" * 50)

    return S, G

S_final, G_final = candidate_elimination(X, y)
print("\nFinal Specific boundary S:\n", S_final)
print("\nFinal General boundary G:\n", G_final)
```
